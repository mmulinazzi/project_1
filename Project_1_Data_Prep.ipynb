```python
import pandas as pd
import numpy as np

#Using a 2GB datafile pulled down via Amazon AWS - Athena
#Search criteria was review data, all columns
#and filtered where product_name = 'toys'
file = 'C:/Users/ahumb/Downloads/2570d538-b194-42d0-80b6-511d4e3a8fc4.csv'

#setting the csv reader
df = pd.read_csv(file)
#df.head()
```

```python
#Running a quick count
df.count()
```

```python
#Dropping the empties
df=df.dropna(how='any')
```

```python
#Running another count to confirm we're good to go
df.count()
```

```python
#Showing the columns and their data types
df.dtypes
```

```python
#Taking a look at the first 5 rows
df.head()
```

```python
#Due to file size (2GB) and GitHub file size limits (of 25MB)
#We decided to pull just LEGO data and perform some international comparisons

#Setting a new dataframe where the product title contains 'LEGO'
#Also resetting the index since we're in a new dataframe
lego_df = pd.DataFrame(df[df['product_title'].str.match('LEGO')])
lego_df = lego_df.reset_index()
del lego_df['index']

```

```python
#When leaving in the body of the reviews, the output file ends up being around 52MB
#By removing reviews and product_type the output file ends up being 22MB

#sets new dataframe and previews the first 5 rows
lego_no_review_df = lego_df.filter(['marketplace','customer_id','review_id','product_id','product_parent','product_title',
                         'star_rating','helpful_votes','total_votes','vine','verified_purchase','review_headline',
                         'review_date','year'],axis=1)
lego_no_review_df.head()
```

```python
#This outputs the file locally to Andy's computer
#File will be uploaded to GitHub manually
lego_no_review_df.to_csv('C:/Users/ahumb/Downloads/lego_no_review_df.csv', header=True,sep=',')
lego_no_review_df.head()
```

```python
#We wanted to store the review work somewhere and created a data frame specific for reviews
#Since review_id is unique we decided to keep it so that we could potentially merge it back for analysis
lego_review_only_df = lego_df.filter(['review_id','review_body'],axis=1)
lego_review_only_df.head()
```

```python
#This outputs the reviews locally to Andy's computer
#Unfortunately file size is an issue again at around 33MB
lego_review_only_df.to_csv('C:/Users/ahumb/Downloads/lego_review_only_df.csv', header=True,sep=',')
lego_review_only_df.head()
```
